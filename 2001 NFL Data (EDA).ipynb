# <p style="padding:10px;background-color:#C8032B;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;overflow:hidden;font-weight:500">2001 Super Bowl Rosters Analysis</p>

```python
def calculate_age_from_dt(birth_date, ref_date=None):
    """
    Calculate age based on a given birth_date and reference date.
    
    Parameters:
        - birth_date (datetime): Date of birth as datetime object.
        - ref_date (str or datetime, optional): The reference date as 'YYYY-MM-DD' or datetime object. 
                                                Defaults to today's date if not provided.
    
    Returns:
        - int: Age in years.
    """
    
    # If ref_date is not provided or is None, use today's date
    if ref_date is None:
        ref_date = datetime.today()
    # If ref_date is provided as a string, convert it to datetime
    elif isinstance(ref_date, str):
        ref_date = datetime.strptime(ref_date, '%Y-%m-%d')
    
    # Calculate the age
    age = ref_date.year - birth_date.year - ((ref_date.month, ref_date.day) < (birth_date.month, birth_date.day))
    
    return age
```

```python
# converting the birth_date column to datetime
NFL_roster_01_slice['birth_date'] = pd.to_datetime(NFL_roster_01_slice['birth_date'])
# Calculate age for the entire column using a custom reference date
custom_date = '2002-02-03'
NFL_roster_01_slice['age'] = NFL_roster_01_slice['birth_date'].apply(lambda x: calculate_age_from_dt(x, custom_date))
#let's look at the average age, height & weight for the Super Bowl Champs
NFL_roster_01_avg = NFL_roster_01_slice[['age','height','weight']].mean()
print(f"The average age for the NFL is: {NFL_roster_01_avg['age']:.2f} years old")
print(f"The average height for the NFL is: {NFL_roster_01_avg['height']:.2f} inches")
print(f"The average weight for the NFL is: {NFL_roster_01_avg['weight']:.2f} pounds")
```
